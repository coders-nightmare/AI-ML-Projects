## Data Preprocessing

### Handiling NULL values

```python
# Release Season, End Year has many null values and does have much importance in recommendation system
```

```python
anime_df.drop(['Release_season','End_year'],axis=1,inplace=True)
```

```python
anime_df.tail()
```

```python
anime_df.info()
```

```python
# replacing Content_warning, Related_manga, Related_anime, Voice_actors, staff 'NaN' values with empty string
```

```python
anime_df[['Content_Warning','Related_Mange','Related_anime','Voice_actors','staff']]=anime_df[['Content_Warning','Related_Mange','Related_anime','Voice_actors','staff']].fillna('')
```

```python
anime_df.info()
```

```python
# as Tags,description , realese_year has less null values,therefore deleting that row
```

```python
anime_df1=anime_df.dropna(subset=['Description','Tags','Release_year'])
anime_df1.info()
```

```python
anime_df1['Studio']=anime_df1['Studio'].fillna('')
anime_df1.info()
```

```python
# Replacing Japanese_name column Nan values Name column values
```

```python
anime_df1['Japanese_name']=anime_df1['Japanese_name'].fillna(anime_df1['Name'][anime_df1['Japanese_name'].isna()])
```

```python
anime_df1.info()
```

```python
# Replacing NaN values of Episodes for Movie type to 1.0
```

```python
anime_df1['Episodes'][anime_df1['Type']=='Movie']=1.0
```

```python
anime_df1.info()
```

```python
anime_df1['Type'].value_counts()
```

```python
# Episodes for music will always be 0
```

```python
anime_df1['Episodes'][anime_df1['Type']=='Music']=0.0
```
