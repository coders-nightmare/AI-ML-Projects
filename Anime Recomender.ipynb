## Data Preprocessing

### Handiling NULL values

```python
# Release Season, End Year has many null values and does have much importance in recommendation system
```

```python
anime_df.drop(['Release_season','End_year'],axis=1,inplace=True)
```

```python
anime_df.tail()
```

```python
anime_df.info()
```

```python
# replacing Content_warning, Related_manga, Related_anime, Voice_actors, staff 'NaN' values with empty string
```

```python
anime_df[['Content_Warning','Related_Mange','Related_anime','Voice_actors','staff']]=anime_df[['Content_Warning','Related_Mange','Related_anime','Voice_actors','staff']].fillna('')
```

```python
anime_df.info()
```

```python
# as Tags,description , realese_year has less null values,therefore deleting that row
```

```python
anime_df1=anime_df.dropna(subset=['Description','Tags','Release_year'])
anime_df1.info()
```

```python
anime_df1['Studio']=anime_df1['Studio'].fillna('')
anime_df1.info()
```

```python
# Replacing Japanese_name column Nan values Name column values
```

```python
anime_df1['Japanese_name']=anime_df1['Japanese_name'].fillna(anime_df1['Name'][anime_df1['Japanese_name'].isna()])
```

```python
anime_df1.info()
```

```python
# Replacing NaN values of Episodes for Movie type to 1.0
```

```python
anime_df1['Episodes'][anime_df1['Type']=='Movie']=1.0
```

```python
anime_df1.info()
```

```python
anime_df1['Type'].value_counts()
```

```python
# Episodes for music will always be 0
```

```python
anime_df1['Episodes'][anime_df1['Type']=='Music']=0.0
```

```python
anime_df1[anime_df1['Episodes'].isna()]
```

```python
anime_df1.info()
```

## One hot Encoding Categorical Values

```python
anime_df1['Type'].value_counts().keys()
```

```python
anime_df1['Type']=anime_df1['Type'].str.strip()
```

```python
df=anime_df1[['Name','Type','Episodes','Tags','Rating','Content_Warning']]
df = df.reset_index(drop=True)
df.tail()
```

```python
df.shape
```

```python
df.Tags
```

```python
[x.strip() for x in df['Tags'][0].split(',')]
```

```python
for idx,tag in enumerate(df['Tags']):
    df['Tags'][idx]=','.join([x.strip() for x in tag.split(',')])
```

```python
df['Tags'][1]
```

```python
df_Tags=df['Tags'].str.get_dummies(sep=',')
```

```python
df_Tags.shape
```

```python
anime_df2=pd.concat([df,df_Tags],axis=1)
```

```python
anime_df2.head()
```

```python
anime_df2.drop(['Tags'],axis=1,inplace=True)
```

```python
anime_df2.head()
```

```python
warning_list=set()
for idx,warning in enumerate(df['Content_Warning']):
#     df['Tags'][idx]=','.join([x.strip() for x in tag.split(',')])
    warning_list.update([x.strip() for x in warning.split(',')])
warning_list
    
```

```python
# Content Warning is covered in tags, therefore we can drop that
anime_df2.drop(['Content_Warning'],axis=1,inplace=True)
```

```python
anime_df2.head()
```

```python
df_type=pd.get_dummies(anime_df2['Type'])
```

```python
anime_df3=pd.concat([anime_df2,df_type],axis=1)
```

```python
anime_df3.drop(['Type'],axis=1,inplace=True)
```
